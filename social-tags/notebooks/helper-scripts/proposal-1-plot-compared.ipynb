```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# delicious
delicious_scores = [
    0.7574578469520103,0.7055025898956535,
    0.6647739004749122,0.6334350358600468,
    0.6067481093659104,0.5827237609928495,
    0.5605666722316297,0.5414560910109746,
    0.5231805724725944,0.506676827684154
]
movielens_scores = [
    0.38782051282051283,0.3358147229114971,
    0.3049705139005897,0.2851054762412103,
    0.27058621282448,0.26026805015131865,
    0.2515207945375543,0.24125874125874125,
    0.23129517483883572,0.22418358340688438
]
```

```python
ks = range(1, 11)

plt.clf()

ax = plt.gca()

ax.set_xlim(0, 11)
ax.set_ylim(0, 0.5)

ax.set_xlabel('k', fontsize=13)
ax.set_ylabel('Micro-averaged F1-score at k', fontsize=13)

ax.plot(ks,delicious_scores,'bo-',label='Delicious T-140')
ax.plot(ks,movielens_scores,'ro-',label="Movielens+IMDB")

ax.set_xticks(np.arange(0, 11, 1))
ax.set_yticks(np.arange(0, 1.05, 0.1))
ax.grid(True)

gridlines = ax.get_xgridlines() + ax.get_ygridlines()
for line in gridlines:
    line.set_linestyle(':')
    line.set_linewidth(0.7)

for k in ks:
    x = k
    y = delicious_scores[k-1]
    label = "{:.2f}".format(y)
    ax.annotate(label,(x,y),xytext=(0, 15), textcoords="offset points",ha='center',va='top')
    

for k in ks:
    x = k
    y = movielens_scores[k-1]
    label = "{:.2f}".format(y)
    ax.annotate(label,(x,y),xytext=(0, 15), textcoords="offset points",ha='center',va='top')
    
ax.legend()

plt.show()
```
